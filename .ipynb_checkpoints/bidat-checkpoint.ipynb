````python
# ============================================================
# CELL 2 — IMPORT LIBRARY
# ============================================================
import pandas as pd
import numpy as np
import pymysql
from sqlalchemy import create_engine, text
from pymongo import MongoClient
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['font.family'] = 'DejaVu Sans'
sns.set_theme(style='whitegrid')

print('Import library berhasil!')
````

````python
# ============================================================
# CELL 3 — LOAD DATASET (INGESTION)
# ============================================================
FILE_PATH = 'all_months_clean.csv'

df = pd.read_csv(FILE_PATH, sep=';')

print('═' * 50)
print('  INFO DATASET')
print('═' * 50)
print(f'  Jumlah baris  : {len(df):,}')
print(f'  Jumlah kolom  : {df.shape[1]}')
print(f'  Ukuran memory : {df.memory_usage(deep=True).sum() / 1024**2:.1f} MB')
print('  Kolom:')
for col in df.columns:
    print(f'    - {col}')
print()
print('Sampel 5 baris pertama:')
display(df.head(5))
````

````python
# ============================================================
# CELL 4 — EKSPLORASI AWAL DATA
# ============================================================
print('Tipe Data:')
print(df.dtypes)
print()
print('Missing Values:')
print(df.isnull().sum())
print()
print('Statistik Deskriptif (Kolom Numerik):')
display(df.describe())
print()
print('Distribusi Kolom Kategorik:')
print('  Status Pesanan:')
print(df['Status Pesanan'].value_counts())
print()
print('  Metode Pembayaran:')
print(df['Metode Pembayaran'].value_counts())
print()
print('  Top 10 Provinsi:')
print(df['Provinsi'].value_counts().head(10))
````

````python
# ============================================================
# CELL 4B — STATISTIK AWAL DATASET (Tambahan Eksplorasi)
# ============================================================
print('═' * 55)
print('  STATISTIK AWAL DATASET')
print('═' * 55)

# Statistik numerik lengkap
print('\n[1] Statistik Deskriptif Lengkap:')
display(df.describe(include='all').T)

# Jumlah missing per kolom dalam persen
print('\n[2] Persentase Missing Value per Kolom:')
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})
display(missing_df[missing_df['missing_count'] > 0])

# Cek duplikasi
dup_count = df.duplicated(subset='order_id').sum()
print(f'\n[3] Duplikasi order_id: {dup_count:,} baris ({dup_count/len(df)*100:.2f}%)')

# Statistik Total Pembayaran
print('\n[4] Statistik Total Pembayaran:')
print(f'  Min    : Rp {df["Total Pembayaran"].min():>15,.0f}')
print(f'  Q1     : Rp {df["Total Pembayaran"].quantile(0.25):>15,.0f}')
print(f'  Median : Rp {df["Total Pembayaran"].median():>15,.0f}')
print(f'  Q3     : Rp {df["Total Pembayaran"].quantile(0.75):>15,.0f}')
print(f'  Max    : Rp {df["Total Pembayaran"].max():>15,.0f}')
print(f'  Mean   : Rp {df["Total Pembayaran"].mean():>15,.0f}')
print(f'  Std    : Rp {df["Total Pembayaran"].std():>15,.0f}')

# Nilai negatif
neg = (df['Total Pembayaran'] < 0).sum()
print(f'\n[5] Nilai Total Pembayaran negatif: {neg:,} baris')
````

````python
# ============================================================
# CELL 5 — DATA CLEANING & PREPROCESSING
# ============================================================
df_clean = df.copy()

# 1. Konversi waktu & ekstrak fitur temporal
df_clean['Waktu Pesanan Dibuat'] = pd.to_datetime(df_clean['Waktu Pesanan Dibuat'])
df_clean['tahun']  = df_clean['Waktu Pesanan Dibuat'].dt.year
df_clean['bulan']  = df_clean['Waktu Pesanan Dibuat'].dt.month
df_clean['hari']   = df_clean['Waktu Pesanan Dibuat'].dt.day_name()
df_clean['jam']    = df_clean['Waktu Pesanan Dibuat'].dt.hour

# 2. Sederhanakan status pesanan
def sederhanakan_status(s):
    if 'Selesai' in str(s): return 'Selesai'
    elif 'Batal'  in str(s): return 'Batal'
    elif 'Dikirim' in str(s): return 'Dalam Pengiriman'
    elif 'diterima' in str(s).lower(): return 'Diterima'
    else: return s

df_clean['status_ringkas'] = df_clean['Status Pesanan'].apply(sederhanakan_status)

# 3. Isi missing values
df_clean['Alasan Pembatalan'] = df_clean['Alasan Pembatalan'].fillna('Tidak Ada')

# 4. Hapus duplikasi order_id
sebelum = len(df_clean)
df_clean = df_clean.drop_duplicates(subset='order_id')
print(f'Duplikasi dihapus: {sebelum - len(df_clean)} baris')

# 5. Hapus total pembayaran negatif
sebelum = len(df_clean)
df_clean = df_clean[df_clean['Total Pembayaran'] >= 0]
print(f'Total pembayaran negatif dihapus: {sebelum - len(df_clean)} baris')

# 6. Normalisasi teks
df_clean['Provinsi']          = df_clean['Provinsi'].str.upper().str.strip()
df_clean['Kota/Kabupaten']    = df_clean['Kota/Kabupaten'].str.upper().str.strip()
df_clean['Metode Pembayaran'] = df_clean['Metode Pembayaran'].str.strip()

# 7. Deteksi outlier (informasi, tidak dihapus)
Q1 = df_clean['Total Pembayaran'].quantile(0.25)
Q3 = df_clean['Total Pembayaran'].quantile(0.75)
IQR = Q3 - Q1
batas_atas = Q3 + 1.5 * IQR
outlier = df_clean[df_clean['Total Pembayaran'] > batas_atas]

print()
print('═' * 50)
print('  HASIL CLEANING')
print('═' * 50)
print(f'  Total baris setelah cleaning : {len(df_clean):,}')
print(f'  Missing values tersisa       : {df_clean.isnull().sum().sum()}')
print(f'  Outlier terdeteksi (tidak dihapus): {len(outlier):,} baris ({len(outlier)/len(df_clean)*100:.1f}%)')
print()
print('  Distribusi Status Pesanan:')
print(df_clean['status_ringkas'].value_counts())
````

````python
# ============================================================
# CELL 6 — FEATURE ENGINEERING
# ============================================================

# Persentase diskon
df_clean['nilai_diskon_persen'] = np.where(
    df_clean['Total Pembayaran'] > 0,
    df_clean['Total Diskon'] / (df_clean['Total Pembayaran'] + df_clean['Total Diskon']) * 100,
    0
).round(2)

# Kuartal
df_clean['kuartal'] = df_clean['bulan'].apply(lambda m: f'Q{(m-1)//3+1}')

# Sesi waktu
df_clean['sesi_waktu'] = df_clean['jam'].apply(
    lambda h: 'Dini Hari' if h < 6 else ('Pagi' if h < 12 else ('Siang' if h < 15 else ('Sore' if h < 18 else 'Malam')))
)

# Multi-kategori flag
df_clean['multi_kategori'] = df_clean['num_product_categories'].apply(
    lambda x: 'Multi Kategori' if x > 1 else 'Single Kategori'
)

# Segmen nilai transaksi
df_clean['segmen_nilai'] = pd.cut(
    df_clean['Total Pembayaran'],
    bins=[0, 100_000, 500_000, 2_000_000, float('inf')],
    labels=['Economy', 'Standard', 'Premium', 'Luxury']
)

print('Feature engineering selesai!')
print('   Fitur baru: nilai_diskon_persen, kuartal, sesi_waktu, multi_kategori, segmen_nilai')
display(df_clean[['order_id', 'kuartal', 'sesi_waktu', 'multi_kategori', 'segmen_nilai', 'nilai_diskon_persen']].head(5))
````

````python
# ============================================================
# CELL 7 — KONEKSI MySQL (XAMPP)
# ============================================================
# Sesuaikan konfigurasi di bawah dengan XAMPP Anda
MYSQL_HOST     = 'localhost'
MYSQL_USER     = 'root'
MYSQL_PASSWORD = ''          # Kosong jika XAMPP default
MYSQL_DB       = 'db_umkm_shopee'
MYSQL_PORT     = 3307

try:
    # Buat database jika belum ada
    conn = pymysql.connect(
        host=MYSQL_HOST, user=MYSQL_USER,
        password=MYSQL_PASSWORD, port=MYSQL_PORT, charset='utf8mb4'
    )
    conn.cursor().execute(
        f'CREATE DATABASE IF NOT EXISTS {MYSQL_DB} '
        f'CHARACTER SET utf8mb4 COLLATE utf8mb4_unicode_ci'
    )
    conn.commit()
    conn.close()

    # Buat SQLAlchemy engine
    engine = create_engine(
        f'mysql+pymysql://{MYSQL_USER}:{MYSQL_PASSWORD}'
        f'@{MYSQL_HOST}:{MYSQL_PORT}/{MYSQL_DB}?charset=utf8mb4'
    )
    with engine.connect() as c:
        c.execute(text('SELECT 1'))

    print(' Koneksi MySQL berhasil!')
    print(f'   Host     : {MYSQL_HOST}:{MYSQL_PORT}')
    print(f'   Database : {MYSQL_DB}')
except Exception as e:
    print(f'Error MySQL: {e}')
    print()
    print('TROUBLESHOOT:')
    print('  1. Pastikan XAMPP → Apache & MySQL sudah START (hijau)')
    print('  2. Cek port MySQL di XAMPP (default 3307)')
    print('  3. Jika password ada, ubah MYSQL_PASSWORD di atas')
    print('  4. Coba buka http://localhost/phpmyadmin untuk verifikasi')
````

````python
# ============================================================
# CELL 8 — STORAGE: UPLOAD DATA KE MySQL
# ============================================================

# --- Tabel 1: transaksi (data utama) ---
kolom_trx = [
    'order_id', 'total_qty', 'total_weight_gr', 'total_returned_qty',
    'Total Diskon', 'num_product_categories', 'status_ringkas',
    'Opsi Pengiriman', 'Metode Pembayaran', 'Kota/Kabupaten', 'Provinsi',
    'Ongkos Kirim Dibayar oleh Pembeli', 'Estimasi Potongan Biaya Pengiriman',
    'Total Pembayaran', 'Perkiraan Ongkos Kirim',
    'Waktu Pesanan Dibuat', 'tahun', 'bulan', 'hari', 'jam',
    'kuartal', 'sesi_waktu', 'multi_kategori', 'nilai_diskon_persen',
    'source_file'
]

df_trx = df_clean[kolom_trx].copy()
df_trx.columns = [
    'order_id', 'total_qty', 'total_weight_gr', 'total_returned_qty',
    'total_diskon', 'num_kategori', 'status_pesanan',
    'opsi_pengiriman', 'metode_pembayaran', 'kota_kabupaten', 'provinsi',
    'ongkir_dibayar_pembeli', 'potongan_ongkir',
    'total_pembayaran', 'perkiraan_ongkir',
    'waktu_pesanan', 'tahun', 'bulan', 'hari', 'jam',
    'kuartal', 'sesi_waktu', 'multi_kategori', 'nilai_diskon_persen',
    'source_file'
]

print(' Upload tabel transaksi...')
df_trx.to_sql('transaksi', con=engine, if_exists='replace', index=False, chunksize=1000)
print(f' Tabel "transaksi" selesai! ({len(df_trx):,} baris)')

# --- Tabel 2: transaksi_kategori (exploded) ---
df_kat = df_clean[['order_id', 'product_categories']].copy()
df_kat['product_categories'] = df_kat['product_categories'].str.split(', ')
df_kat = df_kat.explode('product_categories').reset_index(drop=True)
df_kat.columns = ['order_id', 'kategori']
df_kat['kategori'] = df_kat['kategori'].str.strip()

df_kat.to_sql('transaksi_kategori', con=engine, if_exists='replace', index=False, chunksize=1000)
print(f' Tabel "transaksi_kategori" selesai! ({len(df_kat):,} baris)')

# --- Verifikasi ---
with engine.connect() as conn:
    print()
    print('VERIFIKASI MYSQL:')
    for tbl in ['transaksi', 'transaksi_kategori']:
        c = conn.execute(text(f'SELECT COUNT(*) FROM {tbl}')).fetchone()[0]
        print(f'  {tbl:28s}: {c:>8,} baris')
````

````python
# ============================================================
# CELL 8B — QUERY ANALITIK MYSQL: KATEGORI PRODUK TERLARIS
# ============================================================
print('  QUERY MYSQL: ANALISIS KATEGORI PRODUK')

# Query 1: Top 10 Kategori Terlaris
print('\n[1] Top 10 Kategori Produk Terlaris (MySQL):')
df_mysql_kategori = pd.read_sql("""
    SELECT
        tk.kategori,
        COUNT(DISTINCT tk.order_id)     AS jumlah_order,
        SUM(t.total_pembayaran)         AS total_penjualan,
        ROUND(AVG(t.total_pembayaran),0)AS avg_nilai_order,
        SUM(t.total_qty)                AS total_qty_terjual
    FROM transaksi_kategori tk
    JOIN transaksi t ON tk.order_id = t.order_id
    WHERE t.status_pesanan = 'Selesai'
    GROUP BY tk.kategori
    ORDER BY jumlah_order DESC
    LIMIT 10
""", engine)
display(df_mysql_kategori)

# Query 2: Kategori vs Metode Pembayaran
print('\n[2] Kategori vs Metode Pembayaran (MySQL JOIN):')
df_mysql_kat_bayar = pd.read_sql("""
    SELECT
        tk.kategori,
        t.metode_pembayaran,
        COUNT(*) AS jumlah
    FROM transaksi_kategori tk
    JOIN transaksi t ON tk.order_id = t.order_id
    WHERE t.status_pesanan = 'Selesai'
    GROUP BY tk.kategori, t.metode_pembayaran
    ORDER BY tk.kategori, jumlah DESC
    LIMIT 20
""", engine)
display(df_mysql_kat_bayar)

# Query 3: Analisis Opsi Pengiriman via MySQL
print('\n[3] Analisis Opsi Pengiriman via MySQL:')
df_mysql_pengiriman = pd.read_sql("""
    SELECT
        opsi_pengiriman,
        COUNT(*)                         AS jumlah_order,
        ROUND(AVG(ongkir_dibayar_pembeli), 0) AS avg_ongkir,
        ROUND(AVG(perkiraan_ongkir), 0)       AS avg_perkiraan,
        SUM(ongkir_dibayar_pembeli)           AS total_ongkir,
        ROUND(AVG(potongan_ongkir), 0)        AS avg_potongan
    FROM transaksi
    GROUP BY opsi_pengiriman
    ORDER BY jumlah_order DESC
""", engine)
display(df_mysql_pengiriman)

print(' Query analitik MySQL selesai!')
````

````python
# ============================================================
# CELL 9 — KONEKSI MongoDB (MongoDB Compass)
# ============================================================
# Sesuaikan URI jika Anda menggunakan auth (user:password)
MONGO_URI = 'mongodb://localhost:27017/'
MONGO_DB  = 'db_umkm_shopee'

try:
    client   = MongoClient(MONGO_URI, serverSelectionTimeoutMS=5000)
    client.server_info()   # Trigger koneksi — error jika gagal
    mongo_db = client[MONGO_DB]
    print(' Koneksi MongoDB berhasil!')
    print(f'   URI      : {MONGO_URI}')
    print(f'   Database : {MONGO_DB}')
    print()
    print('   Koleksi yang ada:')
    for col_name in mongo_db.list_collection_names():
        print(f'     - {col_name}')
except Exception as e:
    print(f' Error MongoDB: {e}')
    print()
    print('TROUBLESHOOT:')
    print('  1. Pastikan MongoDB service sudah berjalan:')
    print('     Windows: Services → MongoDB → Start')
    print('     atau jalankan: mongod --dbpath C:\\data\\db')
    print('  2. Cek di MongoDB Compass: connect ke mongodb://localhost:27017')
    print('  3. Jika ada auth, ubah MONGO_URI menjadi:')
    print('     mongodb://username:password@localhost:27017/')
````

````python
# ============================================================
# CELL 10 — STORAGE: UPLOAD DATA KE MongoDB
# ============================================================

def safe_int(val, default=0):
    try:
        return default if pd.isna(val) else int(val)
    except: return default

def safe_str(val, default=''):
    try:
        return default if pd.isna(val) else str(val)
    except: return default

# --- Koleksi 1: detail_transaksi (dokumen nested lengkap) ---
col_detail = mongo_db['detail_transaksi']
col_detail.drop()

docs = []
for _, row in df_clean.iterrows():
    docs.append({
        'order_id':          safe_str(row['order_id']),
        'waktu_pesanan':     str(row['Waktu Pesanan Dibuat']),
        'tahun':             safe_int(row['tahun']),
        'bulan':             safe_int(row['bulan']),
        'kuartal':           safe_str(row['kuartal']),
        'sesi_waktu':        safe_str(row['sesi_waktu']),
        'status':            safe_str(row['status_ringkas']),
        'alasan_pembatalan': safe_str(row['Alasan Pembatalan']),
        'produk': {
            'kategori':        [k.strip() for k in str(row['product_categories']).split(',')],
            'jumlah_kategori': safe_int(row['num_product_categories']),
            'total_qty':       safe_int(row['total_qty']),
            'total_berat_gr':  safe_int(row['total_weight_gr']),
            'multi_kategori':  safe_str(row['multi_kategori']),
        },
        'pembayaran': {
            'metode':             safe_str(row['Metode Pembayaran']),
            'total_diskon':       safe_int(row['Total Diskon']),
            'total_bayar':        safe_int(row['Total Pembayaran']),
            'diskon_persen':      float(row['nilai_diskon_persen']),
            'segmen_nilai':       safe_str(str(row['segmen_nilai'])),
        },
        'pengiriman': {
            'opsi':             safe_str(row['Opsi Pengiriman']),
            'kota_kabupaten':   safe_str(row['Kota/Kabupaten']),
            'provinsi':         safe_str(row['Provinsi']),
            'ongkir_pembeli':   safe_int(row['Ongkos Kirim Dibayar oleh Pembeli']),
            'potongan_ongkir':  safe_int(row['Estimasi Potongan Biaya Pengiriman']),
            'perkiraan_ongkir': safe_int(row['Perkiraan Ongkos Kirim']),
        },
    })

BATCH = 1000
total = 0
for i in range(0, len(docs), BATCH):
    col_detail.insert_many(docs[i:i+BATCH])
    total += len(docs[i:i+BATCH])
    print(f'   ... {total:,} dokumen diunggah', end='\r')
print(f'\n Koleksi "detail_transaksi" selesai! ({total:,} dokumen)')

# --- Koleksi 2: log_pembatalan ---
df_batal = df_clean[df_clean['status_ringkas'] == 'Batal'].copy()
col_batal = mongo_db['log_pembatalan']
col_batal.drop()

docs_batal = []
for _, row in df_batal.iterrows():
    docs_batal.append({
        'order_id':          safe_str(row['order_id']),
        'waktu_pesanan':     str(row['Waktu Pesanan Dibuat']),
        'provinsi':          safe_str(row['Provinsi']),
        'kota_kabupaten':    safe_str(row['Kota/Kabupaten']),
        'kategori_produk':   [k.strip() for k in str(row['product_categories']).split(',')],
        'metode_pembayaran': safe_str(row['Metode Pembayaran']),
        'alasan_pembatalan': safe_str(row['Alasan Pembatalan']),
        'total_qty':         safe_int(row['total_qty']),
        'potensi_revenue':   safe_int(row['Total Pembayaran']),
    })

col_batal.insert_many(docs_batal)
print(f' Koleksi "log_pembatalan" selesai! ({len(docs_batal):,} dokumen)')

# --- Verifikasi MongoDB ---
print()
print('VERIFIKASI MONGODB:')
for col_name in mongo_db.list_collection_names():
    count = mongo_db[col_name].count_documents({})
    print(f'  {col_name:28s}: {count:>8,} dokumen')
````

````python
# ============================================================
# CELL 11 — INTEGRASI SQL ↔ NoSQL (CROSS-SYSTEM ANALYSIS)
# ============================================================
print('═' * 62)
print('  LAYER 1: MySQL Aggregation — Penjualan per Provinsi')
print('═' * 62)

# Layer 1: Query MySQL
sql_layer = pd.read_sql("""
    SELECT
        provinsi,
        COUNT(*)                                          AS total_pesanan,
        SUM(total_pembayaran)                             AS total_revenue,
        COUNT(CASE WHEN status_pesanan = 'Selesai' THEN 1 END) AS pesanan_selesai,
        COUNT(CASE WHEN status_pesanan = 'Batal'   THEN 1 END) AS pesanan_batal,
        ROUND(AVG(total_pembayaran), 0)                   AS avg_nilai_order,
        SUM(total_diskon)                                 AS total_diskon_diberikan
    FROM transaksi
    GROUP BY provinsi
    ORDER BY total_revenue DESC
""", engine)

print(f'  ✓ {len(sql_layer)} provinsi dari MySQL')
display(sql_layer.head(5))

print()
print('═' * 62)
print('  LAYER 2: MongoDB Aggregation — Pembatalan per Provinsi')
print('═' * 62)

# Layer 2: Aggregation MongoDB
mongo_layer = pd.DataFrame(list(mongo_db['log_pembatalan'].aggregate([
    {
        '$group': {
            '_id':               '$provinsi',
            'total_pembatalan':  {'$sum': 1},
            'alasan_terbanyak':  {'$first': '$alasan_pembatalan'},
            'revenue_hilang':    {'$sum': '$potensi_revenue'},
        }
    },
    {'$sort': {'total_pembatalan': -1}},
    {'$project': {
        '_id': 1,
        'total_pembatalan': 1,
        'alasan_terbanyak': 1,
        'revenue_hilang':   1
    }}
])))

if not mongo_layer.empty:
    mongo_layer.rename(columns={'_id': 'provinsi'}, inplace=True)
    print(f'  ✓ {len(mongo_layer)} provinsi dari MongoDB')
    display(mongo_layer.head(5))
else:
    mongo_layer = pd.DataFrame({'provinsi': []})
    print('  ⚠ Belum ada data pembatalan di MongoDB')

print()
print('═' * 62)
print('  LAYER 3: JOIN & ENRICHMENT MySQL + MongoDB')
print('═' * 62)

# Layer 3: Cross-system JOIN
cross_layer = sql_layer.merge(mongo_layer, on='provinsi', how='left')
cross_layer['total_pembatalan'] = cross_layer['total_pembatalan'].fillna(0).astype(int)
cross_layer['revenue_hilang']   = cross_layer['revenue_hilang'].fillna(0).astype(int)
cross_layer['alasan_terbanyak'] = cross_layer['alasan_terbanyak'].fillna('N/A')

cross_layer['rasio_konversi_pct']     = (cross_layer['pesanan_selesai'] / cross_layer['total_pesanan'] * 100).round(1)
cross_layer['tingkat_pembatalan_pct'] = (cross_layer['total_pembatalan'] / cross_layer['total_pesanan'] * 100).round(1)
cross_layer['loss_vs_revenue_pct']    = (cross_layer['revenue_hilang'] / cross_layer['total_revenue'] * 100).round(1)

def risk_category(row):
    lp, cp = row['loss_vs_revenue_pct'], row['tingkat_pembatalan_pct']
    if lp > 20 or cp > 30: return 'CRITICAL'
    elif lp > 10 or cp > 15: return 'HIGH'
    elif lp > 5  or cp > 7:  return 'MEDIUM'
    else: return 'LOW'

cross_layer['risk_level'] = cross_layer.apply(risk_category, axis=1)

display_cols = ['provinsi', 'total_pesanan', 'total_revenue', 'rasio_konversi_pct',
                'total_pembatalan', 'tingkat_pembatalan_pct', 'revenue_hilang', 'risk_level']

print('  ✓ Hasil integrasi (Top 10 Provinsi):')
display(cross_layer[display_cols].head(10))

# ─ SIMPAN KEMBALI KE KEDUA DATABASE ─
print()
print('  Menyimpan hasil integrasi ke MySQL & MongoDB...')

# Ke MySQL
cross_save = cross_layer[display_cols + ['alasan_terbanyak', 'avg_nilai_order']].copy()
cross_save.to_sql('cross_system_analytics', con=engine, if_exists='replace', index=False)
print('  ✓ Tersimpan di MySQL: tabel cross_system_analytics')

# Ke MongoDB
col_cross = mongo_db['cross_system_analytics']
col_cross.drop()
col_cross.insert_many(cross_save.to_dict('records'))
print('  ✓ Tersimpan di MongoDB: koleksi cross_system_analytics')

print()
print('═' * 62)
print('  INSIGHT INTEGRASI:')
worst = cross_layer.sort_values('rasio_konversi_pct').iloc[0]
best  = cross_layer.sort_values('rasio_konversi_pct', ascending=False).iloc[0]
print(f'  Provinsi konversi TERTINGGI : {best["provinsi"]} ({best["rasio_konversi_pct"]}%)')
print(f'  Provinsi konversi TERENDAH  : {worst["provinsi"]} ({worst["rasio_konversi_pct"]}%)')
print(f'  Rata-rata konversi nasional  : {cross_layer["rasio_konversi_pct"].mean():.1f}%')
risk_counts = cross_layer['risk_level'].value_counts()
print(f'  Distribusi Risk Level        : {dict(risk_counts)}')
````

````python
# ============================================================
# ============================================================
# CELL 13 — SPARK: LOAD DATA via CSV (Windows-safe method)
# ============================================================
# Strategi: simpan df_clean ke CSV sementara → Spark baca CSV
# (menghindari createDataFrame crash karena Python worker serialization)

import os
import numpy as np

# ─ Step 1: Siapkan DataFrame yang akan ditulis ke CSV ─
kolom_spark = {
    'order_id':                           'order_id',
    'Waktu Pesanan Dibuat':               'waktu_pesanan',
    'Total Pembayaran':                   'total_pembayaran',
    'Total Diskon':                       'total_diskon',
    'Provinsi':                           'provinsi',
    'Kota/Kabupaten':                     'kota_kabupaten',
    'Metode Pembayaran':                  'metode_pembayaran',
    'Opsi Pengiriman':                    'opsi_pengiriman',
    'Ongkos Kirim Dibayar oleh Pembeli':  'ongkir_pembeli',
    'Perkiraan Ongkos Kirim':             'perkiraan_ongkir',
    'total_qty':                          'total_qty',
    'total_weight_gr':                    'total_weight_gr',
    'status_ringkas':                     'status_ringkas',
    'tahun':                              'tahun',
    'bulan':                              'bulan',
    'jam':                                'jam',
    'sesi_waktu':                         'sesi_waktu',
    'kuartal':                            'kuartal',
    'nilai_diskon_persen':                'nilai_diskon_persen',
}

df_export = df_clean[list(kolom_spark.keys())].rename(columns=kolom_spark).copy()

# ─ Step 2: Bersihkan semua tipe ─
# Categorical → string
for col in df_export.select_dtypes(include='category').columns:
    df_export[col] = df_export[col].astype(str)

# Datetime → string ISO
for col in df_export.select_dtypes(include='datetime64').columns:
    df_export[col] = df_export[col].dt.strftime('%Y-%m-%d %H:%M:%S')

# Float inf/NaN → 0
for col in df_export.select_dtypes(include=['float64','float32']).columns:
    df_export[col] = df_export[col].replace([np.inf, -np.inf], 0).fillna(0)

# Int NaN → 0
for col in ['total_qty','total_weight_gr','tahun','bulan','jam']:
    df_export[col] = df_export[col].fillna(0).astype(int)

# String NaN → ''
for col in df_export.select_dtypes(include='object').columns:
    df_export[col] = df_export[col].fillna('').astype(str)

# ─ Step 3: Tulis ke CSV sementara ─
CSV_TEMP = 'spark_input_temp.csv'
df_export.to_csv(CSV_TEMP, index=False, encoding='utf-8-sig')
print(f' CSV sementara ditulis: {CSV_TEMP} ({os.path.getsize(CSV_TEMP)/1024**2:.1f} MB)')

# ─ Step 4: Spark baca CSV dengan schema eksplisit ─
schema = StructType([
    StructField('order_id',            StringType(),  True),
    StructField('waktu_pesanan',       StringType(),  True),
    StructField('total_pembayaran',    DoubleType(),  True),
    StructField('total_diskon',        DoubleType(),  True),
    StructField('provinsi',            StringType(),  True),
    StructField('kota_kabupaten',      StringType(),  True),
    StructField('metode_pembayaran',   StringType(),  True),
    StructField('opsi_pengiriman',     StringType(),  True),
    StructField('ongkir_pembeli',      DoubleType(),  True),
    StructField('perkiraan_ongkir',    DoubleType(),  True),
    StructField('total_qty',           IntegerType(), True),
    StructField('total_weight_gr',     IntegerType(), True),
    StructField('status_ringkas',      StringType(),  True),
    StructField('tahun',               IntegerType(), True),
    StructField('bulan',               IntegerType(), True),
    StructField('jam',                 IntegerType(), True),
    StructField('sesi_waktu',          StringType(),  True),
    StructField('kuartal',             StringType(),  True),
    StructField('nilai_diskon_persen', DoubleType(),  True),
])

df_spark = (
    spark.read
    .option('header', 'true')
    .option('encoding', 'UTF-8')
    .schema(schema)
    .csv(CSV_TEMP)
    .repartition(4)    # 4 partisi cukup untuk laptop
)

# ─ Step 5: Cache agar tidak re-read CSV setiap query ─
df_spark.cache()

# ─ Step 6: Register Temp View ─
df_spark.createOrReplaceTempView('transaksi_spark')

row_count = df_spark.count()
print(f' Spark DataFrame berhasil dimuat!')
print(f'   Jumlah baris : {row_count:,}')
print(f'   Jumlah kolom : {len(df_spark.columns)}')
print(f'   Partisi      : {df_spark.rdd.getNumPartitions()}')
print(f'   Source       : CSV → Spark (Windows-safe)')
````

````python
# ============================================================
# CELL 14 — APACHE SPARK: SQL ANALYSIS (4 Query Utama)
# ============================================================

# ─ Analisis 1: Total penjualan per provinsi ─
print('[1/4] Total Penjualan per Provinsi (Spark SQL):')
result1 = spark.sql("""
    SELECT
        provinsi,
        COUNT(*)                        AS jumlah_order,
        SUM(total_pembayaran)           AS total_penjualan,
        ROUND(AVG(total_pembayaran), 0) AS rata_rata_order
    FROM transaksi_spark
    WHERE status_ringkas = 'Selesai'
    GROUP BY provinsi
    ORDER BY total_penjualan DESC
    LIMIT 15
""")
result1.show(truncate=False)

# ─ Analisis 2: Tren penjualan per bulan ─
print('[2/4] Tren Penjualan per Bulan (Spark SQL):')
result2 = spark.sql("""
    SELECT
        tahun,
        bulan,
        COUNT(*)                                                      AS jumlah_order,
        SUM(total_pembayaran)                                         AS total_penjualan,
        SUM(CASE WHEN status_ringkas = 'Batal' THEN 1 ELSE 0 END)    AS jumlah_batal
    FROM transaksi_spark
    GROUP BY tahun, bulan
    ORDER BY tahun, bulan
""")
result2.show(30)

# ─ Analisis 3: Distribusi metode pembayaran ─
print('[3/4] Distribusi Metode Pembayaran (Spark SQL):')
result3 = spark.sql("""
    SELECT
        metode_pembayaran,
        COUNT(*)                                               AS jumlah,
        ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(), 2)    AS persen,
        SUM(total_pembayaran)                                  AS total_nilai
    FROM transaksi_spark
    GROUP BY metode_pembayaran
    ORDER BY jumlah DESC
""")
result3.show(truncate=False)

# ─ Analisis 4: Penjualan per sesi waktu ─
print('[4/4] Penjualan per Sesi Waktu (Spark SQL):')
result4 = spark.sql("""
    SELECT
        sesi_waktu,
        COUNT(*)                        AS jumlah_order,
        ROUND(AVG(total_pembayaran), 0) AS avg_nilai_order,
        SUM(total_pembayaran)           AS total_penjualan
    FROM transaksi_spark
    WHERE status_ringkas = 'Selesai'
    GROUP BY sesi_waktu
    ORDER BY jumlah_order DESC
""")
result4.show()

print(' Spark SQL Analysis (4 query) selesai!')
````

````python
# ============================================================
# CELL 15 — APACHE SPARK: ADVANCED (Window + Segmentasi + Churn)
# ============================================================

# ─ [1/3] Window Function: Running Total per Provinsi ─
print('[1/3] Window Function — Running Total per Provinsi:')
result_window = spark.sql("""
    SELECT
        provinsi, tahun, bulan,
        jumlah_order,
        total_penjualan,
        SUM(total_penjualan) OVER (
            PARTITION BY provinsi ORDER BY tahun, bulan
            ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW
        ) AS running_total,
        ROW_NUMBER() OVER (
            PARTITION BY provinsi ORDER BY total_penjualan DESC
        ) AS rank_bulan
    FROM (
        SELECT
            provinsi, tahun, bulan,
            COUNT(*)              AS jumlah_order,
            SUM(total_pembayaran) AS total_penjualan
        FROM transaksi_spark
        GROUP BY provinsi, tahun, bulan
    ) t
    ORDER BY provinsi, tahun, bulan
""")
result_window.show(15, truncate=False)

# ─ [2/3] Customer Segmentation ─
print('[2/3] Customer Segmentation (Premium / Standard / Economy):')
result_segment = spark.sql("""
    SELECT
        CASE
            WHEN total_pembayaran >= 5000000 THEN 'Premium'
            WHEN total_pembayaran >= 2000000 THEN 'Standard'
            ELSE 'Economy'
        END AS customer_segment,
        metode_pembayaran,
        COUNT(*)                        AS jumlah,
        ROUND(AVG(total_pembayaran), 0) AS avg_value,
        COUNT(DISTINCT provinsi)        AS jumlah_provinsi
    FROM transaksi_spark
    WHERE status_ringkas = 'Selesai'
    GROUP BY customer_segment, metode_pembayaran
    ORDER BY customer_segment DESC, jumlah DESC
""")
result_segment.show(truncate=False)

# ─ [3/3] Success Rate per Provinsi (Churn Analysis) ─
print('[3/3] Success Rate per Provinsi (Top 15):')
result_churn = spark.sql("""
    WITH prov AS (
        SELECT
            provinsi,
            COUNT(*) AS total,
            COUNT(CASE WHEN status_ringkas = 'Selesai' THEN 1 END) AS selesai,
            COUNT(CASE WHEN status_ringkas = 'Batal'   THEN 1 END) AS batal
        FROM transaksi_spark
        GROUP BY provinsi
    )
    SELECT
        provinsi, total, selesai, batal,
        ROUND(selesai * 100.0 / total, 1) AS success_rate_pct,
        ROUND(batal   * 100.0 / total, 1) AS cancel_rate_pct,
        CASE
            WHEN selesai * 100.0 / total >= 90 THEN 'EXCELLENT'
            WHEN selesai * 100.0 / total >= 80 THEN 'GOOD'
            WHEN selesai * 100.0 / total >= 70 THEN 'FAIR'
            ELSE 'POOR'
        END AS quality_rating
    FROM prov
    ORDER BY success_rate_pct DESC
    LIMIT 15
""")
result_churn.show(truncate=False)

# ─ Konversi ke Pandas untuk visualisasi ─
df_provinsi   = result1.toPandas()
df_tren       = result2.toPandas()
df_pembayaran = result3.toPandas()
df_sesi       = result4.toPandas()
df_window     = result_window.toPandas()
df_segment    = result_segment.toPandas()
df_churn      = result_churn.toPandas()

print()
print(' Spark Advanced Analysis selesai — semua hasil tersimpan ke Pandas')
````

````python
# CELL 15B — PYSPARK DATAFRAME API (Non-SQL Approach)
# Menunjukkan penggunaan PySpark DataFrame API (.groupBy, .agg, .filter)
# sebagai pelengkap Spark SQL
from pyspark.sql import functions as F
from pyspark.sql.window import Window

print('[1/4] Analisis Opsi Pengiriman — PySpark DataFrame API:')
df_spark_pengiriman = (
    df_spark
    .groupBy('opsi_pengiriman')
    .agg(
        F.count('order_id').alias('jumlah_order'),
        F.round(F.avg('ongkir_pembeli'), 0).alias('rata_rata_ongkir'),
        F.sum('ongkir_pembeli').alias('total_ongkir'),
        F.round(F.avg('perkiraan_ongkir'), 0).alias('rata_rata_perkiraan'),
    )
    .orderBy(F.desc('jumlah_order'))
)
df_spark_pengiriman.show(truncate=False)

print('[2/4] Analisis Pembatalan — PySpark DataFrame API:')
df_spark_batal = (
    df_spark
    .filter(F.col('status_ringkas') == 'Batal')
    .groupBy('provinsi', 'metode_pembayaran')
    .agg(
        F.count('order_id').alias('jumlah_batal'),
        F.sum('total_pembayaran').alias('potensi_revenue_hilang'),
        F.round(F.avg('total_pembayaran'), 0).alias('avg_nilai_batal'),
    )
    .orderBy(F.desc('jumlah_batal'))
)
df_spark_batal.show(15, truncate=False)

print('[3/4] Window Function — Rank Penjualan per Kuartal (DataFrame API):')
window_kuartal = Window.partitionBy('kuartal').orderBy(F.desc('total_penjualan'))
df_spark_kuartal = (
    df_spark
    .filter(F.col('status_ringkas') == 'Selesai')
    .groupBy('kuartal', 'provinsi')
    .agg(F.sum('total_pembayaran').alias('total_penjualan'))
    .withColumn('rank_provinsi', F.rank().over(window_kuartal))
    .filter(F.col('rank_provinsi') <= 5)
    .orderBy('kuartal', 'rank_provinsi')
)
df_spark_kuartal.show(20, truncate=False)

print('[4/4] Statistik Deskriptif via Spark — Total Pembayaran:')
df_spark.select(
    F.count('total_pembayaran').alias('count'),
    F.round(F.mean('total_pembayaran'), 0).alias('mean'),
    F.round(F.stddev('total_pembayaran'), 0).alias('stddev'),
    F.min('total_pembayaran').alias('min'),
    F.expr('percentile(total_pembayaran, 0.25)').alias('Q1'),
    F.expr('percentile(total_pembayaran, 0.50)').alias('median'),
    F.expr('percentile(total_pembayaran, 0.75)').alias('Q3'),
    F.max('total_pembayaran').alias('max'),
).show()

# Simpan hasil ke Pandas untuk laporan
df_spark_pengiriman_pd = df_spark_pengiriman.toPandas()
df_spark_batal_pd      = df_spark_batal.toPandas()
df_spark_kuartal_pd    = df_spark_kuartal.toPandas()

print(' PySpark DataFrame API selesai!')
````

````python
# ============================================================
# CELL 16B — ANALISIS OPSI PENGIRIMAN & RATA-RATA ONGKIR
# ============================================================
print('═' * 60)
print('  ANALISIS OPSI PENGIRIMAN')
print('═' * 60)

# Analisis opsi pengiriman
df_pengiriman = df_clean.groupby('Opsi Pengiriman').agg(
    jumlah_order        = ('order_id', 'count'),
    total_pembayaran    = ('Total Pembayaran', 'sum'),
    rata_rata_ongkir    = ('Ongkos Kirim Dibayar oleh Pembeli', 'mean'),
    total_ongkir        = ('Ongkos Kirim Dibayar oleh Pembeli', 'sum'),
    rata_rata_potongan  = ('Estimasi Potongan Biaya Pengiriman', 'mean'),
    rata_rata_perkiraan = ('Perkiraan Ongkos Kirim', 'mean'),
).sort_values('jumlah_order', ascending=False).reset_index()

df_pengiriman['rata_rata_ongkir']    = df_pengiriman['rata_rata_ongkir'].round(0)
df_pengiriman['rata_rata_potongan']  = df_pengiriman['rata_rata_potongan'].round(0)
df_pengiriman['rata_rata_perkiraan'] = df_pengiriman['rata_rata_perkiraan'].round(0)
df_pengiriman['persen_order']        = (df_pengiriman['jumlah_order'] / df_pengiriman['jumlah_order'].sum() * 100).round(1)

print('\n[1] Distribusi Opsi Pengiriman:')
display(df_pengiriman[['Opsi Pengiriman','jumlah_order','persen_order',
                        'rata_rata_ongkir','rata_rata_potongan','rata_rata_perkiraan']])

# Opsi pengiriman paling diminati
top_pengiriman = df_pengiriman.iloc[0]
print(f'\n  Opsi paling diminati : {top_pengiriman["Opsi Pengiriman"]} ({top_pengiriman["persen_order"]}%)')
print(f'  Rata-rata ongkir     : Rp {top_pengiriman["rata_rata_ongkir"]:,.0f}')

# Analisis ongkir per provinsi (Top 10)
print('\n[2] Rata-Rata Ongkir per Provinsi (Top 10 Termahal):')
df_ongkir_prov = df_clean.groupby('Provinsi').agg(
    rata_ongkir  = ('Ongkos Kirim Dibayar oleh Pembeli', 'mean'),
    total_order  = ('order_id', 'count'),
    total_ongkir = ('Ongkos Kirim Dibayar oleh Pembeli', 'sum'),
).sort_values('rata_ongkir', ascending=False).head(10).reset_index()
df_ongkir_prov['rata_ongkir'] = df_ongkir_prov['rata_ongkir'].round(0)
display(df_ongkir_prov)

# Analisis status pesanan vs opsi pengiriman
print('\n[3] Cross-analysis: Opsi Pengiriman vs Status Pesanan:')
cross_pengiriman = pd.crosstab(
    df_clean['Opsi Pengiriman'],
    df_clean['status_ringkas'],
    margins=True
)
display(cross_pengiriman)

print('\n[4] Korelasi Ongkir dengan Total Pembayaran:')
corr = df_clean[['Ongkos Kirim Dibayar oleh Pembeli','Perkiraan Ongkos Kirim','Total Pembayaran']].corr()
display(corr.round(3))

print('Analisis Opsi Pengiriman selesai!')
````

````python
# ============================================================
# CELL 17 — VISUALISASI (Matplotlib / Seaborn)
# ============================================================

# Chart 1: Top 10 Provinsi berdasarkan Total Penjualan
fig, ax = plt.subplots(figsize=(12, 6))
top10_prov = df_provinsi.head(10)
bars = ax.barh(top10_prov['provinsi'], top10_prov['total_penjualan'] / 1e6,
               color=sns.color_palette('Blues_r', len(top10_prov)))
ax.set_xlabel('Total Penjualan (Juta Rupiah)', fontsize=12)
ax.set_title('Top 10 Provinsi berdasarkan Total Penjualan\n(Transaksi Selesai)', fontsize=14, fontweight='bold')
for bar, val in zip(bars, top10_prov['total_penjualan']):
    ax.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height()/2,
            f'Rp {val/1e6:.1f}M', va='center', fontsize=9)
plt.tight_layout(); plt.show()

# Chart 2: Tren penjualan per bulan
df_tren['periode'] = df_tren['tahun'].astype(str) + '-' + df_tren['bulan'].astype(str).str.zfill(2)
fig, ax1 = plt.subplots(figsize=(14, 5))
ax2 = ax1.twinx()
ax1.bar(df_tren['periode'], df_tren['total_penjualan']/1e6, alpha=0.6, color='steelblue', label='Total Penjualan (Juta Rp)')
ax2.plot(df_tren['periode'], df_tren['jumlah_order'], color='orange', marker='o', linewidth=2, label='Jumlah Order')
ax1.set_xlabel('Periode'); ax1.set_ylabel('Total Penjualan (Juta Rp)', color='steelblue')
ax2.set_ylabel('Jumlah Order', color='orange')
plt.title('Tren Penjualan & Jumlah Order per Bulan', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
ax1.legend(loc='upper left'); ax2.legend(loc='upper right')
plt.tight_layout(); plt.show()

# Chart 3: Status pesanan
status_count = df_clean['status_ringkas'].value_counts()
colors_status = {'Selesai':'#2ecc71','Batal':'#e74c3c','Dalam Pengiriman':'#3498db','Diterima':'#f39c12'}
fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.bar(status_count.index, status_count.values,
              color=[colors_status.get(s, '#95a5a6') for s in status_count.index])
ax.set_ylabel('Jumlah Order'); ax.set_title('Distribusi Status Pesanan', fontsize=14, fontweight='bold')
for bar, val in zip(bars, status_count.values):
    ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+100,
            f'{val:,}\n({val/len(df_clean)*100:.1f}%)', ha='center', va='bottom', fontsize=10)
plt.tight_layout(); plt.show()

# Chart 4: Distribusi order per jam
jam_data = df_clean.groupby('jam')['order_id'].count().reset_index()
jam_data.columns = ['jam', 'jumlah']
colors = ['#d73027' if j in jam_data.nlargest(3,'jumlah')['jam'].values else '#4575b4' for j in jam_data['jam']]
fig, ax = plt.subplots(figsize=(13, 5))
ax.bar(jam_data['jam'], jam_data['jumlah'], color=colors, edgecolor='white', linewidth=0.5)
ax.set_xlabel('Jam (0-23)'); ax.set_ylabel('Jumlah Order')
ax.set_title('Distribusi Order per Jam (Merah = 3 Jam Tersibuk)', fontsize=13, fontweight='bold')
ax.set_xticks(range(24)); plt.tight_layout(); plt.show()
````

````python
# ============================================================
# CELL 17B — VISUALISASI OPSI PENGIRIMAN & PEMBATALAN
# ============================================================

# Chart 5: Opsi Pengiriman - Bar Chart
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Bar: jumlah order per opsi pengiriman
top_pengiriman_viz = df_clean.groupby('Opsi Pengiriman')['order_id'].count().sort_values(ascending=False).head(10)
axes[0].barh(top_pengiriman_viz.index, top_pengiriman_viz.values,
             color=sns.color_palette('Greens_r', len(top_pengiriman_viz)))
axes[0].set_title('Distribusi Opsi Pengiriman\n(Top 10)', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Jumlah Order')
for i, v in enumerate(top_pengiriman_viz.values):
    axes[0].text(v + 10, i, f'{v:,}', va='center', fontsize=8)

# Bar: rata-rata ongkir per opsi pengiriman
avg_ongkir_viz = df_clean.groupby('Opsi Pengiriman')['Ongkos Kirim Dibayar oleh Pembeli'].mean().sort_values(ascending=False).head(10)
axes[1].barh(avg_ongkir_viz.index, avg_ongkir_viz.values,
             color=sns.color_palette('Oranges_r', len(avg_ongkir_viz)))
axes[1].set_title('Rata-rata Ongkir per Opsi Pengiriman\n(Top 10 Termahal)', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Rata-rata Ongkir (Rp)')
for i, v in enumerate(avg_ongkir_viz.values):
    axes[1].text(v + 100, i, f'Rp {v:,.0f}', va='center', fontsize=8)
plt.tight_layout(); plt.show()

# Chart 6: Tingkat Sukses & Pembatalan per Provinsi (Top 15)
df_sukses_prov = df_clean.groupby('Provinsi').agg(
    total=('order_id','count'),
    selesai=('status_ringkas', lambda x: (x=='Selesai').sum()),
    batal=('status_ringkas', lambda x: (x=='Batal').sum()),
).reset_index()
df_sukses_prov['success_rate'] = (df_sukses_prov['selesai']/df_sukses_prov['total']*100).round(1)
df_sukses_prov['cancel_rate']  = (df_sukses_prov['batal']/df_sukses_prov['total']*100).round(1)
df_sukses_prov = df_sukses_prov.sort_values('total', ascending=False).head(15)

fig, ax = plt.subplots(figsize=(13, 6))
x = range(len(df_sukses_prov))
w = 0.35
bars1 = ax.bar([i - w/2 for i in x], df_sukses_prov['success_rate'], w,
               label='Success Rate (%)', color='#2ecc71', alpha=0.85)
bars2 = ax.bar([i + w/2 for i in x], df_sukses_prov['cancel_rate'],  w,
               label='Cancel Rate (%)',  color='#e74c3c', alpha=0.85)
ax.set_xticks(list(x)); ax.set_xticklabels(df_sukses_prov['Provinsi'], rotation=45, ha='right', fontsize=9)
ax.set_ylabel('Rate (%)')
ax.set_title('Tingkat Sukses & Pembatalan Pesanan per Provinsi (Top 15 by Volume)', fontsize=13, fontweight='bold')
ax.legend(); ax.set_ylim(0, 120)
plt.tight_layout(); plt.show()

print(' Visualisasi Opsi Pengiriman & Pembatalan selesai!')
````

````python
# ============================================================
# CELL 18 — INTERACTIVE DASHBOARD (Plotly)
# ============================================================

# KPI
total_order_kpi   = len(df_clean)
total_revenue_kpi = df_clean.loc[df_clean['status_ringkas']=='Selesai','Total Pembayaran'].sum()
success_rate_kpi  = (df_clean['status_ringkas']=='Selesai').sum() / total_order_kpi * 100
cancel_rate_kpi   = (df_clean['status_ringkas']=='Batal').sum() / total_order_kpi * 100
avg_order_kpi     = df_clean.loc[df_clean['status_ringkas']=='Selesai','Total Pembayaran'].mean()

print('═'*60)
print('  KPI DASHBOARD')
print('═'*60)
print(f'  Total Orders      : {total_order_kpi:,}')
print(f'  Total Revenue     : Rp {total_revenue_kpi:,.0f}')
print(f'  Success Rate      : {success_rate_kpi:.1f}%')
print(f'  Cancellation Rate : {cancel_rate_kpi:.1f}%')
print(f'  Avg Order Value   : Rp {avg_order_kpi:,.0f}')
print('═'*60)

# ─ Chart 1: KPI Gauge ─
fig_kpi = go.Figure(go.Indicator(
    mode='number+gauge',
    value=success_rate_kpi,
    title={'text': 'Order Success Rate (%)'},
    gauge={
        'axis': {'range': [0, 100]},
        'bar': {'color': 'darkblue'},
        'steps': [
            {'range': [0,  50], 'color': '#ff6b6b'},
            {'range': [50, 80], 'color': '#ffd93d'},
            {'range': [80,100], 'color': '#6bcf7f'},
        ],
        'threshold': {'line': {'color':'red','width':4}, 'thickness': 0.75, 'value': 90}
    }
))
fig_kpi.update_layout(title_text='KPI: ORDER SUCCESS RATE', height=400)
fig_kpi.show()

# ─ Chart 2: Revenue per Provinsi (Top 15) ─
df_rev = (df_clean[df_clean['status_ringkas']=='Selesai']
          .groupby('Provinsi')['Total Pembayaran'].sum()
          .sort_values(ascending=False).head(15).reset_index())
df_rev.columns = ['provinsi','revenue']
fig2 = px.bar(df_rev, x='revenue', y='provinsi', orientation='h',
              title='TOP 15 PROVINSI — Total Revenue (Transaksi Selesai)',
              labels={'revenue':'Revenue (Rp)', 'provinsi':'Provinsi'},
              color='revenue', color_continuous_scale='Viridis',
              text=df_rev['revenue'].apply(lambda x: f'Rp {x/1e6:.1f}M'))
fig2.update_traces(textposition='auto')
fig2.update_layout(height=600, showlegend=False, plot_bgcolor='rgba(240,240,240,0.5)')
fig2.show()

# ─ Chart 3: Metode Pembayaran (Pie + Bar) ─
fig3 = make_subplots(rows=1, cols=2, specs=[[{'type':'pie'},{'type':'bar'}]],
                     subplot_titles=['Distribusi (%)','Total Nilai Transaksi'])
fig3.add_trace(go.Pie(labels=df_pembayaran['metode_pembayaran'],
                      values=df_pembayaran['jumlah'], hole=0.35,
                      textinfo='label+percent'), row=1, col=1)
fig3.add_trace(go.Bar(x=df_pembayaran['metode_pembayaran'],
                      y=df_pembayaran['total_nilai']/1e6,
                      marker_color='steelblue',
                      text=[f'{v/1e6:.0f}M' for v in df_pembayaran['total_nilai']],
                      textposition='auto'), row=1, col=2)
fig3.update_layout(title_text='ANALISIS METODE PEMBAYARAN', height=450, showlegend=False)
fig3.update_yaxes(title_text='Nilai (Juta Rp)', row=1, col=2)
fig3.show()

# ─ Chart 4: Tren Harian (Order + Revenue) ─
df_daily = (df_clean.assign(tanggal=df_clean['Waktu Pesanan Dibuat'].dt.date)
            .groupby('tanggal').agg(
                jumlah_order=('order_id','count'),
                total_revenue=('Total Pembayaran','sum')
            ).reset_index())
df_daily['tanggal'] = pd.to_datetime(df_daily['tanggal'])

fig4 = make_subplots(specs=[[{'secondary_y': True}]])
fig4.add_trace(go.Scatter(x=df_daily['tanggal'], y=df_daily['jumlah_order'],
                          name='Jumlah Order', line=dict(color='#3498db', width=2)), secondary_y=False)
fig4.add_trace(go.Scatter(x=df_daily['tanggal'], y=df_daily['total_revenue']/1e6,
                          name='Revenue (Juta Rp)', line=dict(color='#e67e22', width=2),
                          fill='tozeroy', fillcolor='rgba(230,126,34,0.1)'), secondary_y=True)
fig4.update_layout(title_text='TREN HARIAN — Order & Revenue', height=400, hovermode='x unified')
fig4.update_yaxes(title_text='Jumlah Order', secondary_y=False)
fig4.update_yaxes(title_text='Revenue (Juta Rp)', secondary_y=True)
fig4.show()

# ─ Chart 5: Top 10 Kategori Terlaris ─
top10_k = top10_kat.reset_index().head(10)
fig5 = px.bar(top10_k, x='jumlah_order', y='kategori', orientation='h',
              title='TOP 10 KATEGORI PRODUK TERLARIS',
              labels={'jumlah_order':'Jumlah Order','kategori':'Kategori'},
              color='total_penjualan', color_continuous_scale='Blues',
              text=top10_k['jumlah_order'].apply(lambda x: f'{x:,}'))
fig5.update_traces(textposition='auto')
fig5.update_layout(height=500, showlegend=False)
fig5.show()

# ─ Chart 6: Risk Level per Provinsi ─
risk_order = ['CRITICAL', 'HIGH', 'MEDIUM', 'LOW']
risk_colors = {'CRITICAL': '#e74c3c', 'HIGH': '#e67e22', 'MEDIUM': '#f1c40f', 'LOW': '#2ecc71'}
fig6 = px.bar(cross_layer.sort_values('risk_level').head(20),
              x='provinsi', y='tingkat_pembatalan_pct', color='risk_level',
              title='RISK LEVEL PEMBATALAN per PROVINSI (Top 20)',
              labels={'tingkat_pembatalan_pct': 'Tingkat Pembatalan (%)', 'provinsi': 'Provinsi'},
              color_discrete_map=risk_colors,
              category_orders={'risk_level': risk_order})
fig6.update_layout(height=500, xaxis_tickangle=-40)
fig6.show()

print(' Interactive Dashboard selesai! (6 chart interaktif)')
````
