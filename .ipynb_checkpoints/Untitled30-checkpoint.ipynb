````python
import pandas as pd
import numpy as np
import pymysql
from sqlalchemy import create_engine, text
from pymongo import MongoClient
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
from pyspark.sql.types import *
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (12, 5)
plt.rcParams['font.family'] = 'DejaVu Sans'
sns.set_theme(style='whitegrid')

print('Import library berhasil !')
````

````python
#load dataset
FILE_PATH = 'all_months_expanded.csv'

df = pd.read_csv(FILE_PATH, sep=';')

print('INFO DATASET')

print(f'  Jumlah baris  : {len(df):,}')
print(f'  Jumlah kolom  : {df.shape[1]}')
print(f'  Ukuran memory : {df.memory_usage(deep=True).sum() / 1024**2:.1f} MB')

print('  Kolom:')
for col in df.columns:
    print(f'    - {col}')
````

````python
#data ingestion
# Konversi waktu
df['Waktu Pesanan Dibuat'] = pd.to_datetime(df['Waktu Pesanan Dibuat'])
df['tahun'] = df['Waktu Pesanan Dibuat'].dt.year
df['bulan'] = df['Waktu Pesanan Dibuat'].dt.month
df['hari']  = df['Waktu Pesanan Dibuat'].dt.day_name()
df['jam']   = df['Waktu Pesanan Dibuat'].dt.hour

# Sederhanakan status pesanan
def sederhanakan_status(s):
    if 'Selesai' in str(s): return 'Selesai'
    elif 'Batal' in str(s): return 'Batal'
    elif 'Dikirim' in str(s): return 'Dalam Pengiriman'
    elif 'diterima' in str(s): return 'Diterima'
    else: return s

df['status_ringkas'] = df['Status Pesanan'].apply(sederhanakan_status)
df['Alasan Pembatalan'] = df['Alasan Pembatalan'].fillna('Tidak Ada')

print('Persiapan data selesai!')
print()
print('Distribusi Status Ringkas:')
print(df['status_ringkas'].value_counts())
````

````python
#ingestion ke mysql
MYSQL_HOST     = 'localhost'
MYSQL_USER     = 'root'
MYSQL_PASSWORD = ''       
MYSQL_DB       = 'db_umkm_shopee'
MYSQL_PORT     = 3306

try:
    conn = pymysql.connect(host=MYSQL_HOST, user=MYSQL_USER,
                           password=MYSQL_PASSWORD, port=MYSQL_PORT, charset='utf8mb4')
    conn.cursor().execute(f'CREATE DATABASE IF NOT EXISTS {MYSQL_DB} CHARACTER SET utf8mb4 COLLATE utf8mb4_unicode_ci')
    conn.commit(); conn.close()
    engine = create_engine(f'mysql+pymysql://{MYSQL_USER}:{MYSQL_PASSWORD}@{MYSQL_HOST}:{MYSQL_PORT}/{MYSQL_DB}?charset=utf8mb4')
    with engine.connect() as c: c.execute(text('SELECT 1'))
    print('Koneksi MySQL berhasil!')
except Exception as e:
    print(f'Error: {e}')
````

````python
# Upload tabel transaksi
kolom = ['order_id','total_qty','total_weight_gr','total_returned_qty',
         'Total Diskon','num_product_categories','status_ringkas',
         'Opsi Pengiriman','Metode Pembayaran','Kota/Kabupaten','Provinsi',
         'Ongkos Kirim Dibayar oleh Pembeli','Estimasi Potongan Biaya Pengiriman',
         'Total Pembayaran','Perkiraan Ongkos Kirim',
         'Waktu Pesanan Dibuat','tahun','bulan','hari','jam','source_file']

df_trx = df[kolom].copy()
df_trx.columns = ['order_id','total_qty','total_weight_gr','total_returned_qty',
                  'total_diskon','num_kategori','status_pesanan',
                  'opsi_pengiriman','metode_pembayaran','kota_kabupaten','provinsi',
                  'ongkir_dibayar_pembeli','potongan_ongkir',
                  'total_pembayaran','perkiraan_ongkir',
                  'waktu_pesanan','tahun','bulan','hari','jam','source_file']

print('Upload tabel transaksi...')
df_trx.to_sql('transaksi', con=engine, if_exists='replace', index=False, chunksize=1000)
print(f'Tabel "transaksi" selesai! ({len(df_trx):,} baris)')
````

````python
# Upload tabel transaksi_kategori (exploded)
df_kat = df[['order_id','product_categories']].copy()
df_kat['product_categories'] = df_kat['product_categories'].str.split(', ')
df_kat = df_kat.explode('product_categories').reset_index(drop=True)
df_kat.columns = ['order_id','kategori']
df_kat['kategori'] = df_kat['kategori'].str.strip()

df_kat.to_sql('transaksi_kategori', con=engine, if_exists='replace', index=False, chunksize=1000)
print(f'Tabel "transaksi_kategori" selesai! ({len(df_kat):,} baris)')
````

````python
# Verifikasi MySQL
with engine.connect() as conn:
    print('VERIFIKASI MYSQL')
    
    for tbl in ['transaksi','transaksi_kategori']:
        c = conn.execute(text(f'SELECT COUNT(*) FROM {tbl}')).fetchone()[0]
        print(f'  {tbl:28s} : {c:>7,} baris')
  
````

````python
# Contoh query MySQL: top kategori terlaris
q = """
    SELECT k.kategori, COUNT(t.order_id) AS jumlah_order, SUM(t.total_pembayaran) AS total_penjualan
    FROM transaksi t JOIN transaksi_kategori k ON t.order_id = k.order_id
    WHERE t.status_pesanan = 'Selesai'
    GROUP BY k.kategori ORDER BY jumlah_order DESC LIMIT 10
"""
display(pd.read_sql(q, engine))
````

````python
#ingestion ke mongoDB
MONGO_URI = 'mongodb://localhost:27017/'
MONGO_DB  = 'db_umkm_shopee'

try:
    client = MongoClient(MONGO_URI, serverSelectionTimeoutMS=5000)
    client.server_info()
    mongo_db = client[MONGO_DB]
    print('Koneksi MongoDB berhasil!')
except Exception as e:
    print(f'Error: {e}')
````

````python
# Upload koleksi detail_transaksi (dokumen nested)
col_detail = mongo_db['detail_transaksi']
col_detail.drop()

def safe_int(val, default=0):
    try:
        if pd.isna(val): return default
        return int(val)
    except:
        return default

def safe_str(val, default=''):
    try:
        if pd.isna(val): return default
        return str(val)
    except:
        return default

docs = []
for _, row in df.iterrows():
    docs.append({
        'order_id':          safe_str(row['order_id']),
        'waktu_pesanan':     str(row['Waktu Pesanan Dibuat']),
        'tahun':             safe_int(row['tahun']),
        'bulan':             safe_int(row['bulan']),
        'status':            safe_str(row['status_ringkas']),
        'alasan_pembatalan': safe_str(row['Alasan Pembatalan']),
        'produk': {
            'kategori':        [k.strip() for k in str(row['product_categories']).split(',')],
            'jumlah_kategori': safe_int(row['num_product_categories']),
            'total_qty':       safe_int(row['total_qty']),
            'total_berat_gr':  safe_int(row['total_weight_gr']),
        },
        'pembayaran': {
            'metode':       safe_str(row['Metode Pembayaran']),
            'total_diskon': safe_int(row['Total Diskon']),
            'total_bayar':  safe_int(row['Total Pembayaran']),
        },
        'pengiriman': {
            'opsi':             safe_str(row['Opsi Pengiriman']),
            'kota_kabupaten':   safe_str(row['Kota/Kabupaten']),
            'provinsi':         safe_str(row['Provinsi']),
            'ongkir_pembeli':   safe_int(row['Ongkos Kirim Dibayar oleh Pembeli']),
            'potongan_ongkir':  safe_int(row['Estimasi Potongan Biaya Pengiriman']),
            'perkiraan_ongkir': safe_int(row['Perkiraan Ongkos Kirim']),
        },
    })

BATCH = 1000
total = 0
for i in range(0, len(docs), BATCH):
    col_detail.insert_many(docs[i:i+BATCH])
    total += len(docs[i:i+BATCH])
    print(f'   ... {total:,} dokumen', end='\r')

print(f'\nKoleksi "detail_transaksi" selesai! ({total:,} dokumen)')
````

````python
# Upload koleksi log_pembatalan
df_batal = df[df['status_ringkas'] == 'Batal'].copy()
col_batal = mongo_db['log_pembatalan']
col_batal.drop()

docs_batal = []
for _, row in df_batal.iterrows():
    docs_batal.append({
        'order_id':          row['order_id'],
        'waktu_pesanan':     str(row['Waktu Pesanan Dibuat']),
        'provinsi':          row['Provinsi'],
        'kota_kabupaten':    row['Kota/Kabupaten'],
        'kategori_produk':   [k.strip() for k in str(row['product_categories']).split(',')],
        'metode_pembayaran': row['Metode Pembayaran'],
        'alasan_pembatalan': row['Alasan Pembatalan'],
        'total_qty':         int(row['total_qty']),
        'potensi_revenue':   int(row['Perkiraan Ongkos Kirim']),
    })

col_batal.insert_many(docs_batal)
print(f' Koleksi "log_pembatalan" selesai! ({len(docs_batal):,} dokumen)')

# Verifikasi MongoDB
print()
print('=' * 52)
print(' VERIFIKASI MONGODB')
print('=' * 52)
for col_name in mongo_db.list_collection_names():
    count = mongo_db[col_name].count_documents({})
    print(f'  {col_name:28s} : {count:>7,} dokumen')
````

````python
#preprocessing 
print('Jumlah missing values:')
print(df.isnull().sum())
print()
print('Jumlah duplikasi:')
print(f'  Duplikat order_id : {df["order_id"].duplicated().sum()}')
print(f'  Duplikat baris    : {df.duplicated().sum()}')
````

````python
print("QUERY : MySQL + MongoDB")

# --- Ambil data dari MySQL ---
df_mysql_lintas = pd.read_sql("""
    SELECT
        provinsi,
        COUNT(*)              AS total_order,
        SUM(total_pembayaran) AS total_penjualan,
        AVG(total_pembayaran) AS avg_order
    FROM transaksi
    WHERE status_pesanan = 'Selesai'
    GROUP BY provinsi
    ORDER BY total_penjualan DESC
""", engine)

print(f"Data MySQL: {len(df_mysql_lintas)} provinsi")

# --- Ambil data dari MongoDB ---
pipeline_mongo = [
    {
        "$group": {
            "_id": "$provinsi",
            "total_batal": {"$sum": 1},
            "alasan_terbanyak": {"$first": "$alasan"}
        }
    },
    {
        "$sort": {
            "total_batal": -1
        }
    }
]

df_mongo_lintas = pd.DataFrame(
    list(mongo_db["log_pembatalan"].aggregate(pipeline_mongo))
)

df_mongo_lintas.rename(
    columns={"_id":"provinsi"},
    inplace=True
)
print(f"Data MongoDB: {len(df_mongo_lintas)} provinsi dengan pembatalan")

# --- Gabungkan MySQL + MongoDB ---
df_lintas = df_mysql_lintas.merge(df_mongo_lintas, on="provinsi", how="left")
df_lintas["total_batal"]     = df_lintas["total_batal"].fillna(0).astype(int)
df_lintas["rasio_batal_pct"] = (df_lintas["total_batal"] / df_lintas["total_order"] * 100).round(1)
df_lintas["total_penjualan"] = df_lintas["total_penjualan"].apply(lambda x: f"Rp {x:,.0f}")
df_lintas["avg_order"]       = df_lintas["avg_order"].apply(lambda x: f"Rp {x:,.0f}")

print()
print("Hasil Gabungan MySQL + MongoDB (Top 10 Provinsi):")
print("-"*60)
display(df_lintas[["provinsi","total_order","total_penjualan",
                    "avg_order","total_batal","rasio_batal_pct",
                    "alasan_terbanyak"]].head(10))

print()
print("INSIGHT:")
worst = df_lintas.sort_values("rasio_batal_pct", ascending=False).iloc[0]
best  = df_lintas.sort_values("rasio_batal_pct").iloc[0]
print(f"  Provinsi dengan pembatalan tertinggi : {worst["provinsi"]} ({worst["rasio_batal_pct"]}%)")
print(f"  Provinsi dengan pembatalan terendah  : {best["provinsi"]} ({best["rasio_batal_pct"]}%)")
print(f"  Rata-rata rasio pembatalan nasional   : {df_lintas["rasio_batal_pct"].mean():.1f}%")
````

````python
# cleaning
df_clean = df.copy()

# 1. Hapus duplikasi
sebelum = len(df_clean)
df_clean = df_clean.drop_duplicates(subset='order_id')
print(f'Hapus duplikasi: {sebelum - len(df_clean)} baris dihapus')

# 2. Isi missing values
df_clean['Alasan Pembatalan'] = df_clean['Alasan Pembatalan'].fillna('Tidak Ada')
print(f'Missing values kolom Alasan Pembatalan diisi "Tidak Ada"')

# 3. Hapus baris total pembayaran negatif
sebelum = len(df_clean)
df_clean = df_clean[df_clean['Total Pembayaran'] >= 0]
print(f'Hapus total pembayaran negatif: {sebelum - len(df_clean)} baris dihapus')

# 4. Normalisasi teks
df_clean['Provinsi']        = df_clean['Provinsi'].str.upper().str.strip()
df_clean['Kota/Kabupaten']  = df_clean['Kota/Kabupaten'].str.upper().str.strip()
df_clean['Metode Pembayaran'] = df_clean['Metode Pembayaran'].str.strip()
print(f'Normalisasi teks selesai')

print()
print(f'Total baris sebelum cleaning : {len(df):,}')
print(f'Total baris setelah cleaning : {len(df_clean):,}')
````

````python
Q1 = df_clean['Total Pembayaran'].quantile(0.25)
Q3 = df_clean['Total Pembayaran'].quantile(0.75)
IQR = Q3 - Q1
batas_atas = Q3 + 1.5 * IQR

outlier = df_clean[df_clean['Total Pembayaran'] > batas_atas]

print(f' Deteksi Outlier Total Pembayaran:')
print(f'   Q1          : Rp {Q1:,.0f}')
print(f'   Q3          : Rp {Q3:,.0f}')
print(f'   IQR         : Rp {IQR:,.0f}')
print(f'   Batas atas  : Rp {batas_atas:,.0f}')
print(f'   Jumlah outlier: {len(outlier):,} baris ({len(outlier)/len(df_clean)*100:.1f}%)')

print(f' Outlier tidak dihapus (bisa transaksi besar yang valid)')
````

````python
# Feature engineering
df_clean['nilai_diskon_persen'] = np.where(
    df_clean['Total Pembayaran'] > 0,
    df_clean['Total Diskon'] / (df_clean['Total Pembayaran'] + df_clean['Total Diskon']) * 100,
    0
).round(2)

df_clean['kuartal'] = df_clean['bulan'].apply(
    lambda m: f'Q{(m-1)//3+1}'
)

df_clean['sesi_waktu'] = df_clean['jam'].apply(
    lambda h: 'Dini Hari' if h < 6 else ('Pagi' if h < 12 else ('Siang' if h < 15 else ('Sore' if h < 18 else 'Malam')))
)

df_clean['multi_kategori'] = df_clean['num_product_categories'].apply(
    lambda x: 'Multi Kategori' if x > 1 else 'Single Kategori'
)

print('Feature engineering selesai!')
print('   Fitur baru: nilai_diskon_persen, kuartal, sesi_waktu, multi_kategori')
display(df_clean[['order_id','kuartal','sesi_waktu','multi_kategori','nilai_diskon_persen']].head(5))
````

# ===== UPGRADED APACHE SPARK: READING FROM DATABASE =====

## Apache Spark akan membaca data langsung dari MySQL menggunakan JDBC connector
### Ini lebih powerful untuk Big Data processing (> 100k records)

````python
# ===== SPARK JDBC READER: MEMBACA LANGSUNG DARI MYSQL =====

# URL JDBC untuk MySQL
JDBC_DRIVER = "com.mysql.cj.jdbc.Driver"
JDBC_URL = f"jdbc:mysql://{MYSQL_HOST}:{MYSQL_PORT}/{MYSQL_DB}?useSSL=false&serverTimezone=UTC"

print("=" * 60)
print("SPARK JDBC: MEMBACA DATA DARI MYSQL")
print("=" * 60)

try:
    # Membaca tabel transaksi dari MySQL
    df_spark_mysql = spark.read \
        .format("jdbc") \
        .option("url", JDBC_URL) \
        .option("dbtable", "transaksi") \
        .option("user", MYSQL_USER) \
        .option("password", MYSQL_PASSWORD) \
        .option("driver", JDBC_DRIVER) \
        .load()
    
    print(f"✓ Berhasil membaca tabel TRANSAKSI dari MySQL")
    print(f"  Jumlah baris: {df_spark_mysql.count():,}")
    print(f"  Jumlah kolom: {len(df_spark_mysql.columns)}")
    
    # Membaca tabel transaksi_kategori dari MySQL
    df_spark_kategori = spark.read \
        .format("jdbc") \
        .option("url", JDBC_URL) \
        .option("dbtable", "transaksi_kategori") \
        .option("user", MYSQL_USER) \
        .option("password", MYSQL_PASSWORD) \
        .option("driver", JDBC_DRIVER) \
        .load()
    
    print(f"\n✓ Berhasil membaca tabel TRANSAKSI_KATEGORI dari MySQL")
    print(f"  Jumlah baris: {df_spark_kategori.count():,}")
    
    print("\n✓ JDBC Reader berhasil diinialisasi!")
    
except Exception as e:
    print(f"⚠ JDBC Reader belum siap: {e}")
    print("  Jika error 'jdbc:mysql driver', install dengan:")
    print("  spark.jars.packages = org.apache.spark:spark-sql-connector-mysql_2.12:1.0.2-2.12")
````

````python
# ===== SPARK MONGODB READER (ALTERNATIVE) =====

# Jika MongoDB sudah tersedia, Spark bisa baca langsung
# Saat ini kita gunakan PyMongo aggregate yang lebih sederhana

print("\n" + "=" * 60)
print("SPARK MONGODB: MEMBACA DATA DARI MONGODB")
print("=" * 60)

# Ambil data dari MongoDB dan convert ke Spark DataFrame
try:
    # Detail Transaksi dari MongoDB
    df_mongo_detail = pd.DataFrame(list(
        mongo_db['detail_transaksi'].find().limit(0)  # Get schema
    ))
    
    # Log Pembatalan dari MongoDB  
    pipeline = [{"$limit": 100}]  # Sample 100 docs untuk schema
    df_mongo_cancel = pd.DataFrame(list(
        mongo_db['log_pembatalan'].aggregate(pipeline)
    ))
    
    # Convert ke Spark DataFrame
    df_spark_mongo_cancel = spark.createDataFrame(df_mongo_cancel)
    
    print(f"✓ Berhasil membaca DETAIL_TRANSAKSI dari MongoDB")
    print(f"✓ Berhasil membaca LOG_PEMBATALAN dari MongoDB")
    print(f"  Jumlah doc pembatalan (sample): {df_mongo_cancel.shape[0]}")
    
except Exception as e:
    print(f"⚠ MongoDB Reader: {e}")
    print("  Spark akan terus gunakan CSV sebagai fallback")
````

````python
# ===== ADVANCED INTEGRATION: SQL ↔ NOSQL (CROSS-SYSTEM ANALYSIS) =====

print("\n" + "=" * 60)
print("ADVANCED INTEGRATION: MYSQL + MONGODB CROSS-SYSTEM ANALYSIS")
print("=" * 60)

# ===== LAYER 1: MYSQL AGGREGATION =====

sql_layer = pd.read_sql("""
    SELECT 
        provinsi,
        COUNT(*) AS total_pesanan,
        SUM(total_pembayaran) AS total_revenue,
        COUNT(CASE WHEN status_pesanan = 'Selesai' THEN 1 END) AS pesanan_selesai,
        COUNT(CASE WHEN status_pesanan = 'Batal' THEN 1 END) AS pesanan_batal,
        ROUND(AVG(total_pembayaran), 0) AS avg_nilai_order,
        COUNT(DISTINCT metode_pembayaran) AS jumlah_metode_unik
    FROM transaksi
    GROUP BY provinsi
    ORDER BY total_revenue DESC
""", engine)

print("\n[1/3] MySQL Aggregation - Penjualan per Provinsi:")
print(f"      {len(sql_layer)} provinsi ditemukan")

# ===== LAYER 2: MONGODB AGGREGATION =====

mongo_layer = pd.DataFrame(list(mongo_db['log_pembatalan'].aggregate([
    {
        "$group": {
            "_id": "$provinsi",
            "total_pembatalan": {"$sum": 1},
            "alasan_paling_sering": {"$first": "$alasan_pembatalan"},
            "revenue_hilang": {"$sum": "$potensi_revenue"}
        }
    },
    {"$sort": {"total_pembatalan": -1}},
    {"$project": {
        "_id": 1,
        "total_pembatalan": 1, 
        "alasan_paling_sering": 1,
        "revenue_hilang": 1
    }}
])))

if not mongo_layer.empty:
    mongo_layer.rename(columns={'_id': 'provinsi'}, inplace=True)
    print("\n[2/3] MongoDB Aggregation - Pembatalan per Provinsi:")
    print(f"      {len(mongo_layer)} provinsi dengan pembatalan ditemukan")
else:
    mongo_layer = pd.DataFrame({'provinsi': []})
    print("\n[2/3] MongoDB Aggregation - Belum ada data pembatalan")

# ===== LAYER 3: CROSS-DATABASE JOIN & ENRICHMENT =====

# Merge hasil MySQL + MongoDB
cross_layer = sql_layer.merge(mongo_layer, on='provinsi', how='left')

# Fill missing values
cross_layer['total_pembatalan'] = cross_layer['total_pembatalan'].fillna(0).astype(int)
cross_layer['revenue_hilang'] = cross_layer['revenue_hilang'].fillna(0).astype(int)
cross_layer['alasan_paling_sering'] = cross_layer['alasan_paling_sering'].fillna('N/A')

# Hitung metrik cross-system
cross_layer['rasio_konversi_pct'] = (
    cross_layer['pesanan_selesai'] / cross_layer['total_pesanan'] * 100
).round(1)

cross_layer['tingkat_pembatalan_pct'] = (
    cross_layer['total_pembatalan'] / cross_layer['total_pesanan'] * 100
).round(1)

cross_layer['loss_vs_revenue_pct'] = (
    cross_layer['revenue_hilang'] / cross_layer['total_revenue'] * 100
).round(1)

# Kategori risk
def risk_category(loss_pct, cancel_pct):
    if loss_pct > 20 or cancel_pct > 30:
        return 'CRITICAL'
    elif loss_pct > 10 or cancel_pct > 15:
        return 'HIGH'
    elif loss_pct > 5 or cancel_pct > 7:
        return 'MEDIUM'
    else:
        return 'LOW'

cross_layer['risk_level'] = cross_layer.apply(
    lambda x: risk_category(x['loss_vs_revenue_pct'], x['tingkat_pembatalan_pct']),
    axis=1
)

print("\n[3/3] Cross-Database Integration Result:")
print(f"      Integrasi selesai dengan {len(cross_layer)} provinsi")
print(f"      Metrik baru: rasio_konversi, tingkat_pembatalan, revenue_loss, risk_level")

print("\n" + "=" * 60)
print("HASIL INTEGRASI SQL ↔ NOSQL (TOP 10 PROVINSI):")
print("=" * 60)

display_cols = ['provinsi', 'total_pesanan', 'total_revenue', 'rasio_konversi_pct',
                'total_pembatalan', 'tingkat_pembatalan_pct', 'revenue_hilang', 
                'loss_vs_revenue_pct', 'risk_level']

display(cross_layer[display_cols].head(10).to_string())

# ===== SAVE KEMBALI KE DATABASE =====

# Save ke MySQL sebagai tabel baru
try:
    cross_layer_sql = cross_layer[[
        'provinsi', 'total_pesanan', 'total_revenue', 'pesanan_selesai', 
        'pesanan_batal', 'rasio_konversi_pct', 'tingkat_pembatalan_pct',
        'revenue_hilang', 'loss_vs_revenue_pct', 'risk_level'
    ]].copy()
    
    cross_layer_sql.to_sql('cross_system_analytics', con=engine, if_exists='replace', index=False)
    print(f"\n✓ Hasil integrasi disimpan ke MySQL tabel: cross_system_analytics")
except Exception as e:
    print(f"⚠ Gagal menyimpan ke MySQL: {e}")

# Save ke MongoDB sebagai koleksi baru
try:
    col_cross = mongo_db['cross_system_analytics']
    col_cross.drop()
    
    records = cross_layer[display_cols].to_dict('records')
    col_cross.insert_many(records)
    print(f"✓ Hasil integrasi disimpan ke MongoDB koleksi: cross_system_analytics")
except Exception as e:
    print(f"⚠ Gagal menyimpan ke MongoDB: {e}")

print("\n✅ ADVANCED INTEGRATION SELESAI!")
````

````python
# ===== SPARK ADVANCED: WINDOW FUNCTIONS & CROSS-SYSTEM JOIN =====

print("\n" + "=" * 60)
print("SPARK ADVANCED: WINDOW FUNCTIONS & ANALYTICS")
print("=" * 60)

from pyspark.sql.window import Window

# ===== ANALISIS 1: RUNNING TOTAL PER PROVINSI (WINDOW FUNCTION) =====

result_window1 = spark.sql("""
    SELECT 
        provinsi,
        tahun,
        bulan,
        jumlah_order,
        total_penjualan,
        SUM(total_penjualan) OVER (PARTITION BY provinsi ORDER BY tahun, bulan) 
            AS running_total_penjualan,
        ROW_NUMBER() OVER (PARTITION BY provinsi ORDER BY total_penjualan DESC) 
            AS rank_bulan_dalam_provinsi
    FROM (
        SELECT 
            provinsi, tahun, bulan,
            COUNT(*) AS jumlah_order,
            SUM(total_pembayaran) AS total_penjualan
        FROM transaksi_spark
        GROUP BY provinsi, tahun, bulan
    )
    ORDER BY provinsi, tahun, bulan
""")

print("\n[1] Window Function - Running Total Penjualan per Provinsi:")
result_window1.show(20, truncate=False)

# ===== ANALISIS 2: PERCENTILE & QUARTILE ANALYSIS =====

result_percentile = spark.sql("""
    SELECT 
        provinsi,
        COUNT(*) AS total_order,
        PERCENTILE_APPROX(total_pembayaran, 0.25) AS quartile_25,
        PERCENTILE_APPROX(total_pembayaran, 0.50) AS median,
        PERCENTILE_APPROX(total_pembayaran, 0.75) AS quartile_75,
        MIN(total_pembayaran) AS min_order,
        MAX(total_pembayaran) AS max_order,
        ROUND(AVG(total_pembayaran), 0) AS avg_order
    FROM transaksi_spark
    WHERE status_ringkas = 'Selesai'
    GROUP BY provinsi
    ORDER BY avg_order DESC
    LIMIT 10
""")

print("\n[2] Percentile Analysis - Order Value Distribution per Provinsi (Top 10):")
result_percentile.show(truncate=False)

# ===== ANALISIS 3: CUSTOMER SEGMENTATION BY PURCHASE PATTERN =====

result_segment = spark.sql("""
    SELECT 
        CASE 
            WHEN total_pembayaran >= 5000000 THEN 'Premium'
            WHEN total_pembayaran >= 2000000 THEN 'Standard'
            ELSE 'Economy'
        END AS customer_segment,
        metode_pembayaran,
        COUNT(*) AS jumlah_customer,
        ROUND(AVG(total_pembayaran), 0) AS avg_value,
        COUNT(DISTINCT provinsi) AS provinsi_terjangkau
    FROM transaksi_spark
    WHERE status_ringkas = 'Selesai'
    GROUP BY customer_segment, metode_pembayaran
    ORDER BY customer_segment DESC, jumlah_customer DESC
""")

print("\n[3] Customer Segmentation - Premium/Standard/Economy by Payment Method:")
result_segment.show(truncate=False)

# ===== ANALISIS 4: CHURN ANALYSIS (ORDER SUCCESS RATE) =====

result_churn = spark.sql("""
    WITH province_metrics AS (
        SELECT 
            provinsi,
            COUNT(*) AS total_order,
            COUNT(CASE WHEN status_ringkas = 'Selesai' THEN 1 END) AS completed,
            COUNT(CASE WHEN status_ringkas = 'Batal' THEN 1 END) AS cancelled
        FROM transaksi_spark
        GROUP BY provinsi
    )
    SELECT 
        provinsi,
        total_order,
        completed,
        cancelled,
        ROUND(completed * 100.0 / total_order, 1) AS success_rate_pct,
        ROUND(cancelled * 100.0 / total_order, 1) AS cancellation_rate_pct,
        CASE
            WHEN completed * 100.0 / total_order >= 90 THEN 'EXCELLENT'
            WHEN completed * 100.0 / total_order >= 80 THEN 'GOOD'
            WHEN completed * 100.0 / total_order >= 70 THEN 'FAIR'
            ELSE 'POOR'
        END AS quality_rating
    FROM province_metrics
    ORDER BY success_rate_pct DESC
    LIMIT 15
""")

print("\n[4] Churn/Quality Analysis - Success Rate per Provinsi (Top 15):")
result_churn.show(truncate=False)

# ===== CONVERT KE PANDAS UNTUK INSIGHTS =====

df_window1 = result_window1.toPandas()
df_percentile = result_percentile.toPandas()
df_segment = result_segment.toPandas()
df_churn = result_churn.toPandas()

print("\n✅ SPARK ADVANCED ANALYSIS SELESAI - Data siap untuk visualisasi")
````

````python
df_spark = spark.read \
    .option('header','true') \
    .option('inferSchema','true') \
    .option('sep',';') \
    .csv('all_months_expanded.csv')

print('Load berhasil!')
print(df_spark.columns)
````

````python
# Rename + register view

df_spark = df_spark.withColumnRenamed('Total Pembayaran', 'total_pembayaran') \
                   .withColumnRenamed('Total Diskon', 'total_diskon') \
                   .withColumnRenamed('Provinsi', 'provinsi') \
                   .withColumnRenamed('Kota/Kabupaten', 'kota_kabupaten') \
                   .withColumnRenamed('Metode Pembayaran', 'metode_pembayaran') \
                   .withColumnRenamed('Opsi Pengiriman', 'opsi_pengiriman') \
                   .withColumnRenamed('Status Pesanan', 'status_pesanan_asli') \
                   .withColumnRenamed('Alasan Pembatalan', 'alasan_pembatalan') \
                   .withColumnRenamed('Ongkos Kirim Dibayar oleh Pembeli', 'ongkir_pembeli') \
                   .withColumnRenamed('Estimasi Potongan Biaya Pengiriman', 'potongan_ongkir') \
                   .withColumnRenamed('Perkiraan Ongkos Kirim', 'perkiraan_ongkir') \
                   .withColumnRenamed('Waktu Pesanan Dibuat', 'waktu_pesanan')

df_spark.createOrReplaceTempView('transaksi_spark')

print(' Data berhasil dimuat ke Spark!')
print(f'Jumlah baris : {df_spark.count():,}')
print(f'Jumlah kolom : {len(df_spark.columns)}')
print(f'Partisi      : {df_spark.rdd.getNumPartitions()}')
````

````python
from pyspark.sql import functions as F

# Tambah kolom tahun, bulan, jam dari waktu_pesanan
df_spark = df_spark \
    .withColumn('tahun', F.year(F.to_timestamp('waktu_pesanan'))) \
    .withColumn('bulan', F.month(F.to_timestamp('waktu_pesanan'))) \
    .withColumn('jam',   F.hour(F.to_timestamp('waktu_pesanan'))) \
    .withColumn('status_ringkas',
        F.when(F.col('status_pesanan_asli').contains('Selesai'), 'Selesai')
         .when(F.col('status_pesanan_asli').contains('Batal'), 'Batal')
         .when(F.col('status_pesanan_asli').contains('Dikirim'), 'Dalam Pengiriman')
         .otherwise('Lainnya')
    ) \
    .withColumn('sesi_waktu',
        F.when(F.col('jam') < 6, 'Dini Hari')
         .when(F.col('jam') < 12, 'Pagi')
         .when(F.col('jam') < 15, 'Siang')
         .when(F.col('jam') < 18, 'Sore')
         .otherwise('Malam')
    )

# Update temp view
df_spark.createOrReplaceTempView('transaksi_spark')

# Verifikasi kolom
print('Kolom df_spark sekarang:')
print(df_spark.columns)
print()
print('Sample status_ringkas:')
df_spark.groupBy('status_ringkas').count().show()
````

````python
df_spark.createOrReplaceTempView('transaksi_spark')

print('Temp view berhasil dibuat')
````

````python
spark.sql("""

SELECT
    provinsi,
    COUNT(*) AS total_transaksi,
    SUM(total_pembayaran) AS total_penjualan
FROM transaksi_spark
GROUP BY provinsi
ORDER BY total_penjualan DESC
LIMIT 10

""").show()
````

````python
# Spark SQL — Analisis 1: Total penjualan per provinsi
result1 = spark.sql("""
    SELECT provinsi,
           COUNT(*) AS jumlah_order,
           SUM(total_pembayaran) AS total_penjualan,
           ROUND(AVG(total_pembayaran), 0) AS rata_rata_order
    FROM transaksi_spark
    WHERE status_ringkas = 'Selesai'
    GROUP BY provinsi
    ORDER BY total_penjualan DESC
    LIMIT 15
""")
print(' Total Penjualan per Provinsi (Spark SQL):')
result1.show(truncate=False)
````

````python
# Spark SQL — Analisis 2: Tren penjualan per bulan
result2 = spark.sql("""
    SELECT tahun, bulan,
           COUNT(*) AS jumlah_order,
           SUM(total_pembayaran) AS total_penjualan,
           SUM(CASE WHEN status_ringkas='Batal' THEN 1 ELSE 0 END) AS jumlah_batal
    FROM transaksi_spark
    GROUP BY tahun, bulan
    ORDER BY tahun, bulan
""")
print(' Tren Penjualan per Bulan (Spark SQL):')
result2.show(30)
````

````python
# Spark SQL — Analisis 3: Distribusi metode pembayaran
result3 = spark.sql("""
    SELECT metode_pembayaran,
           COUNT(*) AS jumlah,
           ROUND(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER(), 2) AS persen,
           SUM(total_pembayaran) AS total_nilai
    FROM transaksi_spark
    GROUP BY metode_pembayaran
    ORDER BY jumlah DESC
""")
print('Distribusi Metode Pembayaran (Spark SQL):')
result3.show(truncate=False)
````

````python
# Spark SQL — Analisis 4: Penjualan per sesi waktu
result4 = spark.sql("""
    SELECT sesi_waktu,
           COUNT(*) AS jumlah_order,
           ROUND(AVG(total_pembayaran), 0) AS avg_nilai_order
    FROM transaksi_spark
    WHERE status_ringkas = 'Selesai'
    GROUP BY sesi_waktu
    ORDER BY jumlah_order DESC
""")
print(' Penjualan per Sesi Waktu (Spark SQL):')
result4.show()
````

````python
# PySpark DataFrame API — Analisis 5: Opsi pengiriman vs rata-rata ongkir
result5 = df_spark.filter(F.col('status_ringkas') == 'Selesai') \
    .groupBy('opsi_pengiriman') \
    .agg(
        F.count('order_id').alias('jumlah_order'),
        F.round(F.avg('perkiraan_ongkir'), 0).alias('avg_ongkir'),
        F.round(F.avg('total_pembayaran'), 0).alias('avg_total_bayar')
    ) \
    .orderBy(F.desc('jumlah_order'))

print('Opsi Pengiriman vs Rata-rata Ongkir (PySpark API):')
result5.show(truncate=False)
````

````python
# PySpark — Analisis 6: Tingkat konversi & pembatalan per provinsi
result6 = df_spark.groupBy('provinsi') \
    .agg(
        F.count('order_id').alias('total_order'),
        F.sum(F.when(F.col('status_ringkas') == 'Selesai', 1).otherwise(0)).alias('order_selesai'),
        F.sum(F.when(F.col('status_ringkas') == 'Batal', 1).otherwise(0)).alias('order_batal'),
    ) \
    .withColumn('tingkat_sukses_pct',
        F.round(F.col('order_selesai') / F.col('total_order') * 100, 1)) \
    .withColumn('tingkat_batal_pct',
        F.round(F.col('order_batal') / F.col('total_order') * 100, 1)) \
    .orderBy(F.desc('total_order')) \
    .limit(15)

print('Tingkat Sukses & Pembatalan per Provinsi (PySpark):')
result6.show(truncate=False)
````
